```python
import pandas as pd
import numpy as np
import re
import requests
import json
from datetime import datetime
from datetime import timedelta

# set the rows of dataframe to be displayed, e.g 10 (default is 30+30=60 rows)
pd.set_option('display.max_rows', 14)
# set the columns of dataframe to be displayed, e.g 'None' - display all the columns
pd.set_option('display.max_columns', None)
# set the decimal places of number, e.g 2
# pd.set_option('display.precision',5)
```

```python
with open('cities.txt', 'r') as f:
    raw = f.read()
raw
```

```python
l = raw.split(' - ')
l
```

```python
ll = []
for i in l:
    test = i.split('  ')
    if isinstance(test,list):
        ll.extend(test)
    else:
        ll.append(test)
ll
```

```python
lll = []
for i in ll:
    new = re.sub(r'[0-9]+ ', '', i)
    lll.append(new)
lll
```

```python
llll = []
for i in lll:
    new = re.sub(r'\(.*\)', '', i)
    llll.append(new.strip())
llll
```

```python
cities = pd.DataFrame(llll)
cities
```

```python
# initialise the column for AQI data with default as nan
nandf = pd.DataFrame(columns=['status','city_lat','city_lng','time','aqi','co','no2','o3','pm10','pm25','so2'])
data = pd.concat([cities, nandf])
data = data.rename(columns={0:'city'})
data = data[['city','status','city_lat','city_lng','time','aqi','co','no2','o3','pm10','pm25','so2']] # change column order in specific order
data
```

```python
# function to request through API and get response as a JSON file
# require requests & json package
def getJSONR(url): 
    try:
        ua = {'user-agent':'Chrome/71.0.3578.98'}
        r = requests.get(url, headers = ua, timeout=30)
        r.raise_for_status() # raise error if r.status_code != 200
        r.encoding = r.apparent_encoding
        js = json.loads(r.text)
        return js
    except Exception as e:
        return e

# functions to get air quality data from the JSON response of World's Air Pollution API
def getSTATUS(js):
    try:
        status = js['status']
        return status
    except Exception as e:
        return np.nan

def getTIME(js):
    try:
        time = js['data']['time']['s']
        return time
    except Exception as e:
        return np.nan

def getGEO(js):
    try:
        geo = []
        lat = js['data']['city']['geo'][0]
        lng = js['data']['city']['geo'][1]
        geo.append(lat)
        geo.append(lng)
        return geo
    except Exception as e:
        geo = [np.nan, np.nan]
        return geo
    
def getAQI(js):
    try:
        aqi = js['data']['aqi']
        return aqi
    except Exception as e:
        return np.nan

def getPM25(js):
    try:
        pm25 = js['data']['iaqi']['pm25']['v']
        return pm25
    except Exception as e:
        return np.nan

def getPM10(js):
    try:
        pm10 = js['data']['iaqi']['pm10']['v']
        return pm10
    except Exception as e:
        return np.nan

def getCO(js):
    try:
        co = js['data']['iaqi']['co']['v']
        return co
    except Exception as e:
        return np.nan

def getNO2(js):
    try:
        no2 = js['data']['iaqi']['no2']['v']
        return no2
    except Exception as e:
        return np.nan

def getSO2(js):
    try:
        so2 = js['data']['iaqi']['so2']['v']
        return so2
    except Exception as e:
        return np.nan

def getO3(js):
    try:
        o3 = js['data']['iaqi']['o3']['v']
        return o3
    except Exception as e:
        return np.nan
```

```python
for i in range(0,data.shape[0]):
    try:
        url = 'https://api.waqi.info/feed/'\
            +data.loc[i,'city']\
            +'/?token=c4103eeb230f7e9cd437f3ea4a2fe03216563895'
        js = getJSONR(url)
        # if getSTATUS(js) returns np.nan, re-request the API until it returns valid response
        while not isinstance(getSTATUS(js),str):
            js = getJSONR(url)
        data.loc[i,'status'] = getSTATUS(js)
        data.loc[i,'time'] = getTIME(js)
        data.loc[i,'city_lat'] = getGEO(js)[0]
        data.loc[i,'city_lng'] = getGEO(js)[1]
        data.loc[i,'aqi'] = getAQI(js)
        data.loc[i,'pm25'] = getPM25(js)
        data.loc[i,'pm10'] = getPM10(js)
        data.loc[i,'co'] = getCO(js)
        data.loc[i,'no2'] = getNO2(js)
        data.loc[i,'so2'] = getSO2(js)
        data.loc[i,'o3'] = getO3(js)
    except Exception as e:
        print(e)
        continue
```
